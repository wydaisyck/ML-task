```python
# after flatten, the index of the min element
print(np.argmin(a1))
print(np.argmax(a1))
print(np.min(a1))
print(np.max(a1))
print(np.mean(a1))
print(np.sum(a1))
print(np.std(a1))


# 点积
print(np.dot(a1, a1.T))
print(np.square(a1))
print(np.sqrt(a1))
print(np.abs(a1))

# e^1
print(np.exp(a1))

# >0 return1 =0 return0 <0 return-1
print(np.sign(a1))

print(np.mod(a1, 2))
```

### 7. Matplotlib 

(a) Create two one dimensional arrays x and y and plot y vs x, add title, xlabel, ylabel, grid.

```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 10)
y = np.array([j ** 2 for j in x])
plt.plot(x, y)
plt.title("this is a test")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

(b) Create multiple arrays and plot them with diﬀerent styles, add legends, add text/mathematical equations on the plot.

```python
p1, = plt.plot([1,2,3])
p2, = plt.plot([3,2,1])
l1 = plt.legend([p2, p1], ["line 2", "line 1"], loc='upper left')
plt.text(1, 1,  'text something')
plt.show()
```

(c) Create multiple subplots, play around with the ﬁgure size, text font/size.

```python
# plot with various axes scales
plt.figure()
hfont = {'fontname':'Comic Sans MS'}

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear', **hfont)
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.4, wspace=0.35)

plt.show()
```

(d) Get familiar with get current axis (gca) handle to do the above tasks
(e) Change the limits on x and y axes, use logarithmic axes to plot.

```python
fig, axes = plt.subplots(2,2)

ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]


ax1.plot(x, y)
ax1.set_yscale('log')
ax1.grid(True)
ax1.set_xlabel('logx')
ax1.set_ylabel('logy')

ax2.plot(x, y)
ax2.grid(True)

ax3.plot(x, y)
ax3.set_yscale('symlog')
ax3.grid(True)


# question e
ax4.plot([1,2,3], [1,4,100])
ax4.set_yscale('log')
ax4.set_xscale('log')
ax4.grid(True)

plt.show()
```
