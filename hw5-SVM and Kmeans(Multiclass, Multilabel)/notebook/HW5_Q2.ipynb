#### Monte-Carlo Simulation: Perform the following procedures 50 times, and report the average and standard deviation of the 50 Hamming Distances that you calculate.

#### (a) Use k-means clustering on the whole Anuran Calls (MFCCs) Data Set (do not split the data into train and test, as we are not performing supervised learning in this exercise). Choose k âˆˆ { 1, 2, . . . , 50 } automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any other method you know.

I find the best k using the largest silhouette score

#### (b) In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species.

For every iteration, I output the dataframe showing the cluster's labels

```python
# load the data and packages
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt
import seaborn as sns

raw_data = pd.read_csv('../data/Anuran Calls (MFCCs)/Frogs_MFCCs.csv')

x_data = raw_data.iloc[:,:-4]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def ham_loss(y_real,y_pred):
    return np.sum(np.not_equal(y_real.values, y_pred.values))/float(y_real.values.size)

def ave_ham_distance(y_real,y_pred):
    return np.sum(np.not_equal(y_real.values, y_pred.values))/float(y_real.values.shape[0])


# random_state 50 times
hl_list = []
hd_list = []
hs_list = []
best_k_list = []

for rs in range(0,50):
    ss_list = []
    for i in range(1, 51):
        km = KMeans(n_clusters = i, random_state=rs)
        km.fit(x_data)
        lable_result = km.labels_
        if i != 1:
            ss = silhouette_score(x_data, lable_result, metric='euclidean')
            ss_list.append(ss)
    best_k = np.argwhere(ss_list == np.amax(ss_list)).flatten().tolist()[0] + 2
    best_k_list.append(best_k)
    km_b = KMeans(n_clusters = best_k, random_state=rs)
    km_b.fit(x_data)
    label_predict = km_b.labels_
    y = raw_data.iloc[:,-4:-1]
    y["cluster"] = label_predict
    df_b = pd.DataFrame()
    
    for class_name in ["Family", "Genus", "Species"]:
        y_des = y[class_name].groupby(y["cluster"]).describe()
        df_b[class_name] = y_des["top"]
        
    print("Iteration %s" %rs)
    display(df_b)
    
    for i in df_b.index.tolist():
        f_new = df_b.iloc[i]["Family"]
        g_new = df_b.iloc[i]["Genus"]
        s_new = df_b.iloc[i]["Species"]
        y.loc[y['cluster']==i,'Family_pre'] = f_new
        y.loc[y['cluster']==i,'Genus_pre'] = g_new
        y.loc[y['cluster']==i,'Species_new'] = s_new
    
    y_real = y.iloc[:,0:3]
    y_pred = y.iloc[:,-3:]

    hl = ham_loss(y_real, y_pred)
    hl_list.append(hl)
    hs_list.append(1-hl)
    hd_list.append(ave_ham_distance(y_real, y_pred))
    
    print("\n")
    
```

#### (c) Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance, Hamming score, and Hamming loss between the true labels and the labels assigned by clusters.

I calculate estimated indexes of the 50 iterations in one dataframe.

```python
res = pd.DataFrame({"Monte-Carlo Simulation":list(range(1,51)), "Best_K":best_k_list, 
                    "hamming loss": hl_list, "hamming score": hs_list, "average hamming distance": hd_list})
res.set_index('Monte-Carlo Simulation',inplace=True)
display(res)
```
